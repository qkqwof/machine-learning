```python
import os
import pandas as pd
import numpy as np
import statsmodels.api as sm
```

```python
# 데이터 불러오기
boston = pd.read_csv('C:/Users/User/OneDrive/바탕 화면/qkqwof/머신러닝/data/Boston_house.csv')
boston
```

```python
#종속변수만 불러오기
target = boston[['Target']]
target
```

### crim, rm, lstat, b, tax, age, zn, nox, indus 변수를 통한 다중선형회귀분석

```python
## bostan data에서 원하는 변수만 뽑아오기 
x_data = boston[['CRIM',"RM","LSTAT","B","TAX",'AGE',"ZN",'NOX','INDUS']]
x_data
```

```python
# 상수항추가
x_data1 = sm.add_constant(x_data,has_constant="add")
x_data1
```

```python
# 회귀모델 적합
multi_model1 = sm.OLS(target,x_data1)
fitted_multi_model1 = multi_model1.fit()
```

```python
# 결과 출력
fitted_multi_model1.summary()
```

```python
# full모델의 회귀계수
fitted_multi_model1.params
```

```python
# base모델과 full모델의 잔차비교 
import matplotlib.pyplot as plt
fitted_multi_model1.resid.plot(label="full")
plt.legend()
```

### VIF를 통한 다중공선성 확인 

```python
from statsmodels.stats.outliers_influence import variance_inflation_factor
vif= pd.DataFrame()
vif['VIF'] = [variance_inflation_factor(x_data.values,i) for i in range(x_data.shape[1])]
vif['name'] = x_data.columns
vif
```

```python
## nox 변수 제거후(X_data2) VIF 확인 

vif = pd.DataFrame()
x_data2= x_data.drop('NOX',axis=1)
vif["VIF Factor"] = [variance_inflation_factor(
    x_data2.values, i) for i in range(x_data2.shape[1])]
vif["features"] = x_data2.columns
vif
```

```python
## RM 변수 제거후(x_data3) VIF 확인 

vif = pd.DataFrame()
x_data3= x_data2.drop('RM',axis=1)
vif["VIF Factor"] = [variance_inflation_factor(
    x_data3.values, i) for i in range(x_data3.shape[1])]
vif["features"] = x_data3.columns
vif
```

```python
# nox 변수 제거한 데이터(x_data3) 상수항 추가 후 회귀 모델 적합
# nox, rm 변수 제거한 데이터(x_data4) 상수항 추가 후 회귀 모델 적합
x_data2_ = sm.add_constant(x_data2,has_constant='add')
x_data3_ = sm.add_constant(x_data3,has_constant='add')
model_vif = sm.OLS(target,x_data2_)
fitted_model_vif = model_vif.fit()
model_vif2 = sm.OLS(target,x_data3_)
fitted_model_vif2 = model_vif2.fit()
```

```python
## 회귀모델 결과 비교 
fitted_model_vif.summary()
```
