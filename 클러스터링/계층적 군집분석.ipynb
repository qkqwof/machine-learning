```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
from sklearn import datasets
```

### 계층적 군집분석(iris 데이터)

```python
iris=datasets.load_iris()
```

```python
labels = pd.DataFrame(iris.target)
labels.columns=['labels']
data = pd.DataFrame(iris.data)
data.columns = ['Sepal length','Sepal width','Petal length', 'Petal width']
data = pd.concat([data,labels], axis=1)
```

```python
from scipy.cluster.hierarchy import dendrogram, linkage

linked = linkage(data,'complete')
plt.figure(figsize=(40,20))
dendrogram(linked, 
           labels=labels['labels'].values,
           leaf_rotation=90,
           leaf_font_size=20)
plt.show()
```

```python
from scipy.cluster.hierarchy import fcluster
```

```python
predict = pd.DataFrame(fcluster(linked,3,criterion='distance'))
predict.columns=['predict']
ct = pd.crosstab(predict['predict'],labels['labels'])
print(ct)
```

```python
data.head()
```

```python
data['cluster'] = predict
data
```

```python
plt.scatter(x=data['Sepal length'], y = data['Petal length'],
            c=data['cluster'])
```
